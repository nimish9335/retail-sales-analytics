```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/SampleSuperstore.csv")
```

```python
df.head()
```

```python
df_clean = df.copy()
```

```python
df_clean.head()
```

```python
print("Original :", df.shape)
print("Copy     :", df_clean.shape)
```

```python
df_clean.duplicated().sum()
```

```python
duplicates = df_clean[df_clean.duplicated()]
duplicates.head()
```

```python
duplicates.shape
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean.duplicated().sum()
```

```python
print("Original Shape :", df.shape)
print("Cleaned Shape  :", df_clean.shape)
```

```python
df_clean.to_csv("../data/cleaned/SampleSuperstore_Cleaned.csv", index=False)
```
