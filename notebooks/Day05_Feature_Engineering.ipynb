```python
import pandas as pd
```

```python
df = pd.read_csv("../data/cleaned/SampleSuperstore_Cleaned_Final.csv")
```

```python
feature_df = df.copy()
```

```python
feature_df.head()
```

```python
feature_df.shape
```

```python
feature_df["Profit Margin (%)"] = (
    feature_df["Profit"] / feature_df["Sales"]
) * 100
```

```python
feature_df.head()
```

```python
feature_df[
    ["Sales", "Profit", "Profit Margin (%)"]
].head(10)
```

```python
feature_df["Profit Margin (%)"] = (
    feature_df["Profit Margin (%)"]
    .round(2)
)
```

```python
feature_df[
    ["Sales", "Profit", "Profit Margin (%)"]
].head(10)
```

```python
feature_df["Average Selling Price"] = (
    feature_df["Sales"] / feature_df["Quantity"]
)
```

```python
feature_df["Average Selling Price"] = (
    feature_df["Average Selling Price"]
    .round(2)
)
```

```python
feature_df[
    ["Sales", "Quantity", "Average Selling Price"]
].head(10)
```

```python
feature_df["Average Selling Price"].describe()
```

```python
feature_df["Discount Percentage"] = (
    feature_df["Discount"] * 100
)
```

```python
feature_df["Discount Percentage"] = (
    feature_df["Discount Percentage"]
    .round(2)
)
```

```python
feature_df[
    ["Discount", "Discount Percentage"]
].head(10)
```

```python
feature_df["Discount Percentage"].describe()
```

```python
feature_df[
    [
        "Profit Margin (%)",
        "Average Selling Price",
        "Discount Percentage"
    ]
].head(10)
```

```python
feature_df[
    [
        "Profit Margin (%)",
        "Average Selling Price",
        "Discount Percentage"
    ]
].describe()
```

```python
feature_df[
    [
        "Profit Margin (%)",
        "Average Selling Price",
        "Discount Percentage"
    ]
].isnull().sum()
```

```python
feature_df.head()
```

```python
feature_df.to_csv(
    "../data/cleaned/SampleSuperstore_Feature_Engineered.csv",
    index=False
)
```
