# What are the overall business KPIs that summarize the retail sales performance?

```python
kpi_summary = {
    "Total Sales": round(df["Sales"].sum(), 2),
    "Total Profit": round(df["Profit"].sum(), 2),
    "Total Quantity Sold": df["Quantity"].sum(),
    "Average Sales per Order": round(df["Sales"].mean(), 2),
    "Average Profit per Order": round(df["Profit"].mean(), 2),
    "Average Discount": round(df["Discount"].mean(), 2)
}

kpi_df = pd.DataFrame(
    kpi_summary.items(),
    columns=["KPI", "Value"]
)

kpi_df
```

# 📈 KPI Analysis

## 💰 1. Total Sales

**Definition:**

Represents the total revenue generated from all customer orders.

**Business Importance:**

- Measures the overall business size.
- Indicates revenue generation capability.
- Used to evaluate business growth over time.

**Business Insight:**

Higher total sales indicate strong market demand and effective sales performance.

---

## 💵 2. Total Profit

**Definition:**

Represents the total profit earned after accounting for business transactions.

**Business Importance:**

- Measures financial success.
- Indicates whether the business is operating profitably.
- Helps evaluate overall business sustainability.

**Business Insight:**

A profitable business can invest in expansion, marketing, and product development.

---

## 🛒 3. Total Quantity Sold

**Definition:**

Represents the total number of products sold across all orders.

**Business Importance:**

- Evaluates product demand.
- Supports inventory planning.
- Assists supply chain management.

**Business Insight:**

Higher product sales indicate strong market demand and effective inventory utilization.

---

## 📊 4. Average Sales per Order

**Definition:**

Represents the average revenue generated from a single order.

**Business Importance:**

- Measures customer spending behavior.
- Evaluates pricing effectiveness.
- Helps identify opportunities for upselling and cross-selling.

**Business Insight:**

Increasing the average order value directly contributes to higher overall revenue.

---

## 💹 5. Average Profit per Order

**Definition:**

Represents the average profit earned from each customer order.

**Business Importance:**

- Measures transaction-level profitability.
- Helps evaluate pricing and cost strategies.
- Supports profit optimization efforts.

**Business Insight:**

Higher average profit per order indicates healthier business margins.

---

## 🏷️ 6. Average Discount

**Definition:**

Represents the average discount offered across all customer orders.

**Business Importance:**

- Evaluates pricing strategy.
- Measures promotional effectiveness.
- Helps balance sales growth with profitability.

**Business Insight:**

Moderate discounts can attract customers, while excessive discounts may reduce profit margins.
