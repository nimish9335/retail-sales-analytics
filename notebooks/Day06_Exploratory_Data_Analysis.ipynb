# Which numerical variables are related to each other?

```python
numerical_df = eda_df.select_dtypes(include=["number"])

numerical_df.head()
```

```python
correlation_matrix = numerical_df.corr()

correlation_matrix
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

Observation: The heatmap shows a moderate positive correlation between Sales and Profit, indicating that higher sales generally lead to higher profit. Discount has a negative correlation with Profit, suggesting that larger discounts tend to reduce profitability. Most other relationships are weak, implying that multiple factors influence business performance.

Observation: The scatter plot shows a generally positive relationship between Sales and Profit, indicating that larger sales often generate higher profit. However, several orders with high sales still result in low or negative profit, suggesting that factors such as discounts or costs affect profitability.

# Business Insights

## 1. Sales Distribution
- Sales are highly right-skewed, with most orders having low sales and a few orders contributing very high revenue.

## 2. Profit Distribution
- While most orders are profitable, several transactions result in negative profit, indicating loss-making sales.

## 3. Category Performance
- Technology and Furniture contribute significantly to overall sales.
- Profitability varies across categories, showing that higher sales do not always result in higher profit.

## 4. Sub-Category Analysis
- Some sub-categories consistently generate strong profits, while others contribute to business losses.
- These underperforming sub-categories require pricing and discount review.

## 5. Customer Segment Analysis
- The Consumer segment contributes the highest sales and profit.
- Different customer segments show different purchasing behaviors, suggesting opportunities for targeted marketing.

## 6. Regional Analysis
- The West region is the strongest performer in both sales and profit.
- The Central region generates relatively lower profit despite reasonable sales, indicating lower profitability.

## 7. State Analysis
- California and New York are among the strongest-performing states in terms of sales and profit.
- These states represent key markets for future business growth.

## 8. City Analysis
- Major cities contribute a significant share of revenue and profit.
- High-performing cities should be prioritized for inventory planning and marketing.

## 9. Discount Analysis
- Higher discounts are generally associated with lower profits.
- Excessive discounting should be reviewed to improve overall profitability.

## 10. Correlation Analysis
- Sales and Profit have a moderate positive relationship.
- Discount has a negative relationship with Profit, highlighting its impact on earnings.

## 11. Outlier Detection
- Sales and Profit contain several outliers representing unusually large transactions.
- These transactions should be analyzed separately because they can significantly influence summary statistics.
